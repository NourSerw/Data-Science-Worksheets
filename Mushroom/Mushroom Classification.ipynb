```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm, tree
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import LinearSVC
```

```python
mushrooms_df = pd.read_csv('F:\\Kaggle Datasets\\mushroom-classification\\mushrooms.csv')
```

```python
mushrooms_df.head()
```

```python
def code_stuff(x):
    return ord(x)
```

```python
mushrooms_df = mushrooms_df.applymap(code_stuff)
```

```python
mushrooms_df
```

```python
mushrooms_df_unique = mushrooms_df['class'].value_counts()
```

```python
mushrooms_df_unique
```

```python
train, test = train_test_split(mushrooms_df, test_size=.3)
```

```python
train_x = train.drop('class', axis=1)
train_y = train['class']

test_x = test.drop('class', axis=1)
test_y = test['class']
```

```python
classifiers = []

model1 = svm.SVC()
classifiers.append(model1)
```

```python
model2 = tree.DecisionTreeClassifier()
classifiers.append(model2)
```

```python
model3 = RandomForestClassifier()
classifiers.append(model3)
```

```python
model4 = GaussianNB()
classifiers.append(model4)
```

```python
model5 = LinearSVC()
classifiers.append(model5)
```

```python
for clf in classifiers:
    clf.fit(train_x, train_y)
    y_pred= clf.predict(test_x)
    acc = accuracy_score(test_y, y_pred)
    print("Accuracy of %s is %s"%(clf, acc))
    cm = confusion_matrix(test_y, y_pred)
    print("Confusion Matrix of %s is %s"%(clf, cm))
    f1 = f1_score(test_y,y_pred,average='weighted')
    print("F1 score of %s is %s"%(clf,f1))
```

```python
for clf in classifiers:
    print("Classifier %s"%(clf))
    print(f"Train accuracy {round(clf.score(train_x,train_y)*100,2)} %")
    print(f"Test accuracy {round(clf.score(test_x,test_y)*100,2)} %")
    print("\n")
```
